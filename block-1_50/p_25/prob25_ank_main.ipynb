# $1000$ Digit Fibonacci Numbers

```python
def fib(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        s = fib(x - 1) + fib(x - 2)
    return s
```

```python
i = 31
while(i >= 0):
    a = fib(i)
    L = [i for i in str(a)]
    if len(L) == 8:
        break
    i = i + 1
    
print(i)
```
