We will use Binet's formula for generating Fibonacci sequence. One can fint the proof here: https://artofproblemsolving.com/wiki/index.php/Binet%27s_Formula. Also, the linear algebraic proof is something awesome.

```python
%%timeit
import math

a = 0
b = math.sqrt(5)
n = 12
L = [3 + i*3 for i in range(n)]
for i in L:
    a = a + int(1 / b * (((1 + b) / 2)**i - ((1 - b) / 2)**i))
print(a)
print(L)
```
